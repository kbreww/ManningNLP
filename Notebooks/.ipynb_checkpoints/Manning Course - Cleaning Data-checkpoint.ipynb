### Using Regular Expressions to Clean Text

```python
##testing the correct method of LaTeX removal

testString = "Thank you!, This equation $y=ax+b$, is very helpful."
pattern = r'\$.*\$'
new = re.sub(pattern, '', testString)
new
```

```python
##applying to dataframe using the regex function and pandas replace string function
tempDF.text = tempDF.text.str.replace(pattern, '')
tempDF
```

```python
##removal of HTML cells in same manner as they occur in between < >
pattern2 = r'\<.*\>'
tempDF.text = tempDF.text.str.replace(pattern2, '')
tempDF
```

```python
##removal of digits
pattern3 = r'\b\d+\b'
tempDF.text = tempDF.text.str.replace(pattern3, '')
tempDF
```

```python
## removal of many special characters. besides "!,.-?"
pattern4 = "(\+|@|&|'|\(|\)|<|>|#|\")"
tempDF.text = tempDF.text.str.replace(pattern4, '')
tempDF
```
